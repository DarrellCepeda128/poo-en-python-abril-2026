# lenguaje python, crear la clase "animal" con los atributos "nombre", "raza" crear el constructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase Animal
class Animal:
    # El constructor (__init__) inicializa los atributos al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para simular el ladrido
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau! 🐶")

    # Función para simular que corre
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente 💨.")

    def Morder(self):
        print(f"{self.nombre} está murdiendo a la vecina")

# --- Creación de los objetos (Instancias) ---

# Creamos el objeto 'pitbull' con datos específicos
pitbull = Animal("Thor", "Pitbull 🙀")

# Creamos el objeto 'chihuahua' con datos específicos
chihuahua = Animal("Pipo", "Chihuahua 😼")

# Creamos un tercer objeto 'perrocallejero' con datos específicos
perrocallejero = Animal("Firulais", "Callejero 🐺")

# --- Ejecución de las funciones ---

print(f"Datos del primer perro: {pitbull.nombre} ({pitbull.raza})")
pitbull.ladrar()
pitbull.correr()

print("-" * 30)

print(f"Datos del segundo perro: {chihuahua.nombre} ({chihuahua.raza})")
chihuahua.ladrar()
chihuahua.correr()

print("-" * 30)

print(f"Datos del tercer perro: {perrocallejero.nombre}, Raza: {perrocallejero.raza}")
perrocallejero.ladrar()
perrocallejero.correr()

print("-" * 30)

print("Programa creado por Cepeda Darrell 0030 ✨")
```
